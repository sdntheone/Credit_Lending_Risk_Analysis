```python
# VIF for numerical columns
numeric_columns = []
for i in df.columns:
    if df[i].dtype != 'object' and i not in ['PROSPECTID','Approved_Flag']:
        numeric_columns.append(i)
```

```python
len(numeric_columns)
```

```python
# VIF sequentially check

vif_data = df[numeric_columns]
total_columns = vif_data.shape[1]
columns_to_be_kept = []
column_index = 0
threshold = 6
```

```python
X = df[numeric_columns].copy()

# Convert to numpy array once
X_values = X.values
columns = X.columns.tolist()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
while True:
    vif = [variance_inflation_factor(X_values, i) for i in range(X_values.shape[1])]
    max_vif = max(vif)
    
    if max_vif > threshold:
        max_index = vif.index(max_vif)
        print(f"Dropping '{columns[max_index]}' with VIF = {max_vif:.2f}")
        # Drop from both DataFrame and values array
        X_values = np.delete(X_values, max_index, axis=1)
        del columns[max_index]
    else:
        break

# Final list of kept columns
columns_to_be_kept = columns
print("Remaining columns:", columns_to_be_kept)
```

```python
print("Remaining columns:", columns_to_be_kept)
```

```python
# check Anova for columns_to_be_kept 

from scipy.stats import f_oneway

columns_to_be_kept_numerical = []

for i in columns_to_be_kept:
    a = list(df[i])  
    b = list(df['Approved_Flag'])  
    
    group_P1 = [value for value, group in zip(a, b) if group == 'P1']
    group_P2 = [value for value, group in zip(a, b) if group == 'P2']
    group_P3 = [value for value, group in zip(a, b) if group == 'P3']
    group_P4 = [value for value, group in zip(a, b) if group == 'P4']


    f_statistic, p_value = f_oneway(group_P1, group_P2, group_P3, group_P4)

    if p_value <= 0.05:
        columns_to_be_kept_numerical.append(i)
```
