## Hypothesis Testing

**Chi-Square Test**<br>
To check whether the categorical features are associated with output feature or not. It means whether we can keep categorical feature or we should remove it

### Hypothesis Testing for Feature Selection

For each categorical feature, we define the hypotheses as follows:

#### Null Hypothesis (H0):
There is **no relationship** between the feature and the target variable (they are independent).  
- The feature **does not influence** the target.
- It **should be removed** if (H0) is true.

#### Alternative Hypothesis (H1):
There **is a relationship** between the feature and the target variable (they are dependent).  
- The feature **affects** the target.
- It **should be kept** if (H0) is rejected.

```python
# Chi-square test
from scipy.stats import chi2_contingency
for i in ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']:
    chi2, pval, _, _ = chi2_contingency(pd.crosstab(df[i], df['Approved_Flag']))
    print(i, '---', pval)
```

since all the categorical features have p-value less than 0.05, we will accept all the variable

### Multicollinearity

Multicollinearity occurs when **two or more independent variables** in a regression model are **highly correlated** with each other. This makes it difficult to determine the **individual effect** of each variable on the target.

---

### Why is Multicollinearity a Problem?
1. **Unstable Coefficients**  The regression model struggles to assign proper weights to each feature.
2. **Inconsistent Results**  Small changes in data can lead to large variations in coefficients.
3. **Redundancy**  Some features provide redundant information, making the model more complex without adding value.

---

### Example of Multicollinearity
Suppose you have a dataset with these variables:
- **House Price (Target)**
- **House Size (sq ft)**
- **Number of Bedrooms**
- **Number of Bathrooms**

Since **larger houses tend to have more bedrooms and bathrooms**, "Number of Bedrooms" and "Number of Bathrooms" are **highly correlated** with "House Size." This can create multicollinearity.

---

### How to Detect Multicollinearity?

#### 1. Correlation Matrix (Heatmap)
If two independent variables have a correlation **greater than 0.8 or 0.9**, multicollinearity might be present

#### 2. Variance Inflation Factor (VIF)

A VIF > 5 (or VIF > 10) suggests high multicollinearity.

### Why Not Just Use Correlation to Detect Multicollinearity?

Correlation helps in detecting pairwise relationships between two variables, but it is not always sufficient for identifying multicollinearity in a regression model. Here's why:

- A correlation matrix (or heatmap) shows only the direct **linear relationship** between two variables at a time.
- However, multicollinearity can be a result of **multiple variables** being jointly correlated, even if no single pair has a high correlation.

- In convex functions, correlation can give misleading values because correlation only captures linear relationships and fails to detect non-linear dependencies.

### Why Correlation is Misleading in Convex Functions?
Correlation assumes a linear relationship between two variables. However, if the relationship is convex or non-linear, the correlation coefficient may be close to zero, even though there is a strong relationship between the variables.


```python
# VIF for numerical columns
numeric_columns = []
for i in df.columns:
    if df[i].dtype != 'object' and i not in ['PROSPECTID','Approved_Flag']:
        numeric_columns.append(i)
```

```python
# Selecting numerical features
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
vif_data = df[numeric_columns].copy()

# Set VIF threshold
vif_threshold = 6
columns_to_be_kept = []
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
while True:
    vif_values = [variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
    vif_df = pd.DataFrame({"Feature": vif_data.columns, "VIF": vif_values})
    
    # Print VIF values
    print(vif_df.sort_values(by="VIF", ascending=False))
    
    max_vif = vif_df["VIF"].max()
    
    if max_vif < vif_threshold:
        break  # Stop if all VIFs are below threshold
    
    # Drop feature with highest VIF
    drop_feature = vif_df.loc[vif_df["VIF"] == max_vif, "Feature"].values[0]
    print(f"Dropping {drop_feature} with VIF: {max_vif}")
    vif_data.drop(columns=[drop_feature], inplace=True)

# Store final features with low VIF
columns_to_be_kept = vif_data.columns.tolist()

print("\nFinal Features with Low VIF:\n", columns_to_be_kept)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder

# Encode 'Approved_Flag' separately
label_encoder = LabelEncoder()
df['Approved_Flag'] = label_encoder.fit_transform(df['Approved_Flag'])  #  Encode separately

# Define preprocessor
preprocessor = ColumnTransformer([
    ('ordinal', OrdinalEncoder(), ['EDUCATION']),  #  Pass as a list
    ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'), 
     ['GENDER', 'last_prod_enq2', 'first_prod_enq2', 'MARITALSTATUS'])
], remainder='passthrough')  # Keep other columns as they are
```

```python
transformed_data = preprocessor.fit_transform(df)
```

```python
feature_names = preprocessor.get_feature_names_out()  # Get transformed feature names
df_transformed = pd.DataFrame(transformed_data, columns=feature_names)
```

```python
df_transformed
```

```python
# 1. Random Forest
y = df_transformed['remainder__Approved_Flag']
x = df_transformed. drop ( ['remainder__Approved_Flag'], axis = 1 )
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, precision_recall_fscore_support
import warnings
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
rf_classifier = RandomForestClassifier(n_estimators = 200, random_state=42)
rf_classifier.fit(x_train, y_train)
y_pred = rf_classifier.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print ()
print(f'Accuracy: {accuracy}')
print ()
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred)


for i, v in enumerate(['p1', 'p2', 'p3', 'p4']):
    print(f"Class {v}:")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
```
